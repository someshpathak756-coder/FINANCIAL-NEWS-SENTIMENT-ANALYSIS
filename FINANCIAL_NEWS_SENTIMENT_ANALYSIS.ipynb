```python
# We are creating a 'pipeline' which is like an automated assembly line
from transformers import pipeline

# This line tells Colab to go get the 'FinBERT' brain specifically
sentiment_model = pipeline("sentiment-analysis", model="ProsusAI/finbert")

print("The Financial Brain is loaded and ready!")
```

```python
# Here are 3 news headlines for the AI to test
news_list = [
    "The stock market is reaching all-time highs today!",
    "The company is facing a massive lawsuit and losing money.",
    "The central bank will meet tomorrow to discuss interest rates."
]

# Tell the AI to read the list and give us the answer
results = sentiment_model(news_list)

# Print the answers
for i in range(len(news_list)):
    print(f"News: {news_list[i]}")
    print(f"Result: {results[i]['label']} (Confidence: {results[i]['score']:.2f})")
    print("-" * 30)
```

```python
import pandas as pd

# 1. Turn our results into a neat table
df = pd.DataFrame(results)

# 2. Add the news headlines to the table so we can see them side-by-side
df['headline'] = news_list

# 3. Clean up the look (move the headline to the front)
df = df[['headline', 'label', 'score']]

# 4. Show the final table
print("--- FINAL PROJECT RESULTS ---")
df
```
